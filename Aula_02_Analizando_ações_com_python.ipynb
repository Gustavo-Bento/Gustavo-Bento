# Analizando Grafico por periodo

```python
vale_fatiado = vale[vale.index.year == 2024]
vale_fatiado["Close"].plot(figsize=(22,8),label="IBOV")
vale_fatiado["Close"].rolling(21).mean().plot(label="MM21")
vale_fatiado["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```

# Analizando gráfico por intervalo de tempo

```python
vale_fatiado = vale[(vale.index.year >= 2020)&(vale.index.year <= 2024)]
vale_fatiado["Close"].plot(figsize=(22,8),label="IBOV")
vale_fatiado["Close"].rolling(21).mean().plot(label="MM21")
vale_fatiado["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```
